# <center> Preparation of the data, turn all features numerical </center>

## Load data

```python
from pandas import read_csv
import pandas as pd
train_full = read_csv('../HousePrices/trainfull.csv', index_col=0)
test = read_csv('../HousePrices/test.csv', index_col=0)
```

## Select only the categorical features

```python
train_full_object = train_full.select_dtypes(include='object')
train_full_object = train_full_object.fillna('None')
test_object = test.select_dtypes(include='object')
test_object = test_object.fillna('None')
```

```python
# pip install category_encoders
```

```python
import category_encoders as ce
from sklearn.preprocessing import OneHotEncoder
ohe = ce.OneHotEncoder(handle_unknown='ignore', use_cat_names=True)

trainfull_ohe = ohe.fit_transform(train_full_object)
trainfull_ohe.head()
```

```python
test_ohe = ohe.transform(test_object)
test_ohe.head()
```

## Same treatment for the numerical features

```python
train_full_number = train_full.select_dtypes(exclude='object')
train_full_number = train_full_number.fillna(0)
test_number = test.select_dtypes(exclude='object')
test_number = test_number.fillna(0)
```

Concatener the numerical and categorical features into one dataframe trainfull

```python
import pandas as pd
trainfull = pd.concat([train_full_number, trainfull_ohe], axis=1)
trainfull.head()
```

```python
test = pd.concat([test_number, test_ohe], axis=1)
test.head()
```

## Split data

Let's do 80-20%:

```python
from numpy import random
SEED = 42
random.seed(SEED)
```

```python
from sklearn.model_selection import train_test_split
VAL_SIZE = 0.2
train, val = train_test_split(trainfull, test_size=VAL_SIZE)
```

```python
print("Train full shape: " + str(trainfull.shape))
print("Train shape: " + str(train.shape))
print("Val shape: " + str(val.shape))
print("Test shape" + str(test.shape))
```

## Save data

```python
trainfull.to_csv("trainfull_ohe.csv")
test.to_csv("test_ohe.csv")
train.to_csv("train_ohe.csv")
val.to_csv("val_ohe.csv")
```
